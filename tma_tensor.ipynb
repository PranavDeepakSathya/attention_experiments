```python
#essentially, we want a mask that can take any nbfloat16 value and give us GMxGN co-rodinate
import torch
```

```python
M = 8192
N = 8192
BM = 128
BN = 128
GM = M//BM
GN = N//BN
print(GM)
print(GN)

x = torch.randn(1,dtype=torch.bfloat16)

def co_ordinate_mask(x,GM,GN,BM,BN): 
  prime = 1171
  shift = 3719
  print(x)
  bits = x.view(torch.uint16)[0].item()
  print(bits)
  t = (prime*bits + shift) % (GM*GN)
  m = (t//GN)*BM
  n = (t%GN)*BN
  return m,n

```

```python
co_ordinate_mask(x,GM,GN,BM,BN)
```
